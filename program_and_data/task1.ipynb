# Task 1-3:计算每台售货机每月的每单平均交易额与日均订单量，以表格形式体现在报告中。

```python
days = [31,28,31,31,30,30,31,31,30,31,30,31]

A_avg_trad = []
A_avg_order = []

B_avg_trad = []
B_avg_order = []

C_avg_trad = []
C_avg_order = []

D_avg_trad = []
D_avg_order = []

E_avg_trad = []
E_avg_order = []
for k in range(1,13):
    sell_A = sell_data_A[[i.month == k for i in sell_data_A['支付时间']]]
    A_avg_trad.append(np.mean(sell_A['实际金额']))
    A_avg_order.append(len(sell_A)/days[k-1])
    
    sell_B = sell_data_B[[i.month == k for i in sell_data_B['支付时间']]]
    B_avg_trad.append(np.mean(sell_B['实际金额']))
    B_avg_order.append(len(sell_B)/days[k-1])    
    
    sell_C = sell_data_C[[i.month == k for i in sell_data_C['支付时间']]]
    C_avg_trad.append(np.mean(sell_C['实际金额']))
    C_avg_order.append(len(sell_C)/days[k-1])   
    
    sell_D = sell_data_D[[i.month == k for i in sell_data_D['支付时间']]]
    D_avg_trad.append(np.mean(sell_D['实际金额']))
    D_avg_order.append(len(sell_D)/days[k-1])       
    
    sell_E = sell_data_E[[i.month == k for i in sell_data_E['支付时间']]]
    E_avg_trad.append(np.mean(sell_E['实际金额']))
    E_avg_order.append(len(sell_E)/days[k-1])       
```

```python
E_avg_order
```

```python
task1_3 = pd.DataFrame({'月份':range(1,13),'A销售机平均交易额':A_avg_trad,'A销售机日均订单量':A_avg_order,\
                       'B销售机平均交易额':B_avg_trad,'B销售机日均订单量':B_avg_order,\
                       'C销售机平均交易额':C_avg_trad,'C销售机日均订单量':C_avg_order,\
                       'D销售机平均交易额':D_avg_trad,'D销售机日均订单量':D_avg_order,\
                       'E销售机平均交易额':E_avg_trad,'E销售机日均订单量':E_avg_order})
task1_3
```

```python
task1_3.to_csv('task1-3.csv',index = False,encoding = 'gbk')
```
